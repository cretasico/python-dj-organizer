```python
import os
import pandas as pd
from mutagen.mp3 import MP3
from mutagen.easyid3 import EasyID3

# Ruta del archivo Excel
ruta_excel = r'E:\Musica-E\Organizacion\Biblioteca.xlsx'

# Leer el archivo Excel
df = pd.read_excel(ruta_excel)

# Almacenar etiquetas únicas
etiquetas_unicas = set()

# Contador de archivos procesados
contador_canciones = 0
max_canciones = 1000

# Iterar sobre cada carpeta principal en la columna de ubicación
for ubicacion in df['Ubicacion']:
    # Verificar que la ruta base sea una carpeta o archivo existente
    if os.path.exists(ubicacion):
        # Recorrer todas las subcarpetas y archivos
        for root, dirs, files in os.walk(ubicacion):
            for file in files:
                # Verificar límite de canciones
                if contador_canciones >= max_canciones:
                    break

                # Construir la ruta completa del archivo
                file_path = os.path.join(root, file)
                
                # Verificar si el archivo tiene extensión .mp3
                if file_path.lower().endswith('.mp3'):
                    try:
                        # Cargar el archivo MP3
                        audio = MP3(file_path, ID3=EasyID3)
                        # Imprimir el nombre del archivo
                        print(f"Etiquetas en '{file_path}':")
                        # Listar etiquetas y sus valores y agregar etiquetas al conjunto único
                        for etiqueta, valor in audio.items():
                            print(f"  {etiqueta}: {valor}")
                            etiquetas_unicas.add(etiqueta)
                        print("\n")  # Espacio para separar archivos
                        contador_canciones += 1  # Incrementar el contador
                    except Exception as e:
                        print(f"Error al leer {file_path}: {e}")
            if contador_canciones >= max_canciones:
                break
    else:
        print(f"Ignorando {ubicacion}: La ruta no existe o no es válida.")

# Imprimir las etiquetas únicas encontradas en los primeros 300 archivos
print("Etiquetas únicas en los primeros 300 archivos MP3:")
for etiqueta in etiquetas_unicas:
    print(etiqueta)
```
