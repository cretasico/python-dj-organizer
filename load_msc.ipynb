```python
#Cargando todas las canciones al archivo biblioteca
import os
import eyed3
import pandas as pd

def get_mp3_info(file_path):
    try:
        audiofile = eyed3.load(file_path)
        if audiofile is None or audiofile.tag is None:
            # Usar el nombre del archivo sin la extensión si no hay metadatos disponibles
            title = os.path.splitext(os.path.basename(file_path))[0]
            return title, "Unknown", "Unknown", None, None

        # Usa el nombre del archivo sin la extensión si el título no está disponible
        title = audiofile.tag.title if audiofile.tag.title else os.path.splitext(os.path.basename(file_path))[0]
        artist = audiofile.tag.artist if audiofile.tag.artist else "Unknown"
        genre = audiofile.tag.genre.name if audiofile.tag.genre else "Unknown"
        
        # BPM and Key extraction (requires analysis and more advanced libraries)
        bpm = None  # Placeholder for BPM analysis
        key = None  # Placeholder for Key analysis

        return title, artist, genre, bpm, key
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
        return None, None, None, None, None

def scan_music_directory(directory):
    music_data = []
    for root, dirs, files in os.walk(directory):
        for file in files:
            if file.endswith(".mp3"):
                file_path = os.path.join(root, file)
                title, artist, genre, bpm, key = get_mp3_info(file_path)
                if title and artist and genre:
                    music_data.append({
                        "Nombre": os.path.splitext(file)[0],
                        "Titulo": title,
                        "Artista": artist,
                        "Genero": genre,
                        "Ubicacion": root,  # Solo el path sin el nombre del archivo
                        "BPM": bpm,
                        "Clave / Key": key,
                        "Pais": "",
                        "Valoracion": "",
                        "Comentario": ""
                    })
    return music_data

def save_to_excel(data, output_path):
    df = pd.DataFrame(data)
    df.to_excel(output_path, index=False, engine='openpyxl')

# Configuraciones
music_directory = r"E:\Musica-E\DJ Project"
output_file = r"E:\Musica-E\Organizacion\Biblioteca.xlsx"

# Escanear directorio de música y guardar en Excel
music_data = scan_music_directory(music_directory)
save_to_excel(music_data, output_file)

print(f"Archivo Excel creado en: {output_file}")
```
